```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



# Variables to plot
variables = ['Tave_sp','Tave_sm',  'Tave_at','Tave_wt', 'MAT',
             'MCMT', 'MWMT', 'TD','EMT', 'EXT',
             'DD_0', 'DD5',  'DD_18','DD18', 'DD1040',
             'FFP', 'bFFP', 'eFFP','PAS', 'NFFD',
             'Eref', 'CMD', 'CMI','AHM','SHM',
             'PPT_sp','PPT_sm', 'PPT_at', 'PPT_wt','RH',
             'MAP','MSP','barren','cropland', 'forest', 
             'grassland', 'permanent_snow_ice', 'urban', 'water']

# Scenarios
scenarios = ['(1)', '(2)', '(3)', '(4)', '(5)']
colors = ['blue', 'green', 'orange', 'purple', 'red']

# Directory where the CSV file is located
dir_resample_mask = "/Users/liting/Documents/data/resample_mask/"

# Create subplots
fig, axes = plt.subplots(8, 5, figsize=(20, 30))
axes = axes.flatten()
# Load the CSV file
data1 = pd.read_csv(dir_resample_mask + '/historical/mean_sd.csv')
data2 = pd.read_csv(dir_resample_mask + '/ssp126_2071_2100/mean_sd.csv')
data3 = pd.read_csv(dir_resample_mask + '/ssp245_2071_2100/mean_sd.csv')
data4 = pd.read_csv(dir_resample_mask + '/ssp370_2071_2100/mean_sd.csv')
data5 = pd.read_csv(dir_resample_mask + '/ssp585_2071_2100/mean_sd.csv')


for i, variable in enumerate(variables):
    ax = axes[i]
    means = []
    sds = []

    for j,this_data in enumerate([data1,data2,data3,data4,data5]):
        mean_col = "clim_mean"
        sd_col = "clim_sd"
        
        means.append(this_data.loc[this_data['clim_name'] == variable, mean_col].values[0])
        sds.append(this_data.loc[this_data['clim_name'] == variable, sd_col].values[0])
    
    x = np.arange(len(scenarios))
    ax.errorbar(x, means, yerr=sds, fmt='o', label=variable)
    ax.set_xticks(x)
    print(scenarios)
    ax.set_xticklabels(scenarios,rotation=0)
    ax.set_title(variable)
    ax.grid(True)


# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
